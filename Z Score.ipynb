# Z-Score

## Imagine a class where the exam scores are normally distributed with a mean score of 70 and a standard deviation of 10.

## Scenario: 
## You scored 85 on the exam. To understand how well you did relative to your classmates, you can calculate your Z-score.

## Calculate the Z-Score:
## ùëç=(85‚àí70)/10 = 15/10 = 1.5

## Interpretation:

## Z-score of 1.5 means your score is 1.5 standard deviations above the mean.
## In practical terms, you scored better than a significant majority of the class. Using the standard normal distribution table, you can find that a Z-score of 1.5 corresponds approximately to the 93rd percentile. This means you scored higher than about 93% of the students. 

## In this example:

## The mean score (70) corresponds to a Z-score of 0.
## Your score (85) corresponds to a Z-score of 1.5.
## The shaded area under the curve to the right of the Z-score represents the proportion of students who scored more than 85.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define the range of x values
x = np.linspace(-3, 3, 1000)

# Calculate the standard normal distribution
y = stats.norm.pdf(x, 0, 1)

# Plot the standard normal distribution
plt.plot(x, y, label='Standard Normal Distribution')

# Shade the area under the curve for the Z-score
x_shade = np.linspace(1.5, 3, 1000)
y_shade = stats.norm.pdf(x_shade, 0, 1)
plt.fill_between(x_shade, y_shade, alpha=0.5, color='orange', label='Z = 1.5 and above')

# Add annotations
plt.axvline(1.5, color='red', linestyle='--', label='Z = 1.5')
plt.text(1.5, 0.1, '85', horizontalalignment='center', color='red')

# Labels and legend
plt.xlabel('Z-Score')
plt.ylabel('Probability Density')
plt.title('Standard Normal Distribution with Z-Score')
plt.legend()
plt.show()
```
