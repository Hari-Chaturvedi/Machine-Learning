# Learning Simple Linear Regression

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as sc
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

## Loading data

```python
df = pd.read_csv(r'C:\Users\harib\OneDrive\Documents\Python Scripts\Statistics\Regression\Simple linear Regression\placement_2Var.csv')
df
```

## Regrssion Plot

```python
# plt.scatter(df['cgpa'], df['package'])
sns.regplot(x = 'cgpa', y='package', data = df, line_kws={"color":'red'})
plt.show()
```

```python
x = df.iloc[:,0:1]
y = df.iloc[:,-1]
x, y
```

## Spliting Data For Testing and Training 

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2,random_state=2)
```

```python
lr = LinearRegression()
lr
```

```python
lr.fit(x_train, y_train)
```

```python
print(x_test, '\n\n y_test\n\n', y_test)
```

```python
lr.predict(x_test.iloc[0].values.reshape(1,1))
```

### Scatter Plot of Trained Model 


```python
plt.scatter(df['cgpa'], df['package'])
# sns.regplot(x = 'cgpa', y='package', data = df, line_kws={"color":'red'})
plt.plot(x_test, lr.predict(x_test), color='black')
plt.show()
```
